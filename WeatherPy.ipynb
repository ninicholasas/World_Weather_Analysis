```python
# Import Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create a set of random latitude and longtitude combinations.

lats = np.random.uniform(low= -90.000, high= 90.000, size= 1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size= 1500)
lats_lngs = zip(lats, lngs)
lats_lngs
```

```python
# Add the latitudes and longtitudes to a list.
coordinates = list(lats_lngs)
```

```python
# Use the print() function to display the latitude and longtitude combinations.
for coordinate in coordinates:
    print(coordinate[0],coordinate[1])
```

```python
from citipy import citipy
```

```python
# Create a list for holding the cities.

cities = []

# Identify the nearest city for each latitude aand longtitude combination.
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    
    
    # If the city is unique, then we will add it to the cities list.
    if city not in cities:
        cities.append(city)
        
# Print the city count to confirm sufficient count.
len(cities)
```
